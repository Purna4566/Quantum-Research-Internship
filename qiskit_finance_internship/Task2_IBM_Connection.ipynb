```python
from qiskit_ibm_runtime import QiskitRuntimeService

print("Connecting to IBM Quantum Services...")

try:
    # Connect using the newly provisioned instance
    service = QiskitRuntimeService(channel="ibm_quantum_platform")
    print("Authentication Successful!\n")
    
    # Fetch ALL backends you have access to
    all_backends = service.backends()
    
    print("--- Available IBM Quantum Backends ---")
    for backend in all_backends:
        kind = "Simulator" if backend.simulator else "Physical QPU"
        print(f"Name: {backend.name:<20} | Type: {kind:<12} | Qubits: {backend.num_qubits}")
    print("--------------------------------------")
    
except Exception as e:
    print(f"\n[System Error] Connection Failed: {e}")
    
```
