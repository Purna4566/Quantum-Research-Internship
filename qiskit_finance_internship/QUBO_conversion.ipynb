```python
import numpy as np

# Number of assets (let's use 3)
n = 3

# 1. Expected Returns (mu) for Asset 1, 2, and 3
mu = np.array([0.08, 0.09, 0.12])

# 2. Covariance Matrix (Sigma) - How they move together
# Diagonal is individual variance, off-diagonal is covariance
sigma = np.array([
    [0.04, 0.01, 0.02],
    [0.01, 0.05, 0.03],
    [0.02, 0.03, 0.06]
])

# 3. Risk Aversion Factor (theta)
theta = 0.5
# 4. Budget Constraint (k)
k = 2.0
lambda_penalty = 1.0

Q = np.zeros((n, n))    
for i in range(n):
    for j in range(n):
        if i == j:
            Q[i][j] = -mu[i] + (theta * sigma[i][i]) + lambda_penalty * (1 - 2*k)
        else:
                Q[i][j] = theta * sigma[i][j] + lambda_penalty
            
print("----Q matrix for portfolio optimization----")
print(Q)
```
