```python
import numpy as np
import dimod
import neal
Q = np.array([[-3.06 ,  1.005 , 1.01 ],
 [ 1.005, -3.065,  1.015],
 [ 1.01,   1.015, -3.09 ]])

def qubo_to_ising(Q):
    n = len(Q)
    J = np.zeros((n, n))
    h = np.zeros(n)
    offset = 0
    for i in range(n):
        for j in range(n):
            if i == j:
                h[i] += Q[i][j] / 2 
                offset += Q[i][j] / 2
            else:
                J[i][j] += Q[i][j] / 4
                h[i] += Q[i][j] / 4
                h[j] += Q[i][j] / 4
                offset += Q[i][j] / 4
            
    return J, h, offset
    
J_matrix, h_vector, final_offset = qubo_to_ising(Q)
bqm = dimod.BinaryQuadraticModel(h_vector, J_matrix, final_offset, dimod.SPIN)
sampler = neal.SimulatedAnnealingSampler()
sampleset = sampler.sample(bqm, num_reads=1000)
print("Absolute Best Portfolio Found (Ground State):")
print(sampleset.first)
```
