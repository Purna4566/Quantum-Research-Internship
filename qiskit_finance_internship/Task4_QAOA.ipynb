```python
from qiskit_optimization import QuadraticProgram
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit_algorithms import QAOA
from qiskit_algorithms.optimizers import COBYLA
from qiskit.primitives import StatevectorSampler
qc = QuadraticProgram()
qc.binary_var('x')
qc.binary_var('y')
qc.maximize(linear={'x': 2, 'y': 2}, quadratic={('x', 'y'): -4})
print("matheamtical formulation of the problem:")
print(qc.export_as_lp_string())
sampler = StatevectorSampler()
optimizer = COBYLA(maxiter=100)
qoao = QAOA(sampler=sampler, optimizer=optimizer, reps=1)
print("\nExecuting QAOA Hybrid Loop. Please wait...")
qoao_optimizer = MinimumEigenOptimizer(qoao)
result = qoao_optimizer.solve(qc) 
print("\n--- QAOA Optimization Results ---")
print(f"Optimal Variables: x={result.x[0]:.0f}, y={result.x[1]:.0f}")
print(f"Optimal Objective Value: {result.fval}")
print("---------------------------------")
```
