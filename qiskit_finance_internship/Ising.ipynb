```python
import numpy as np
Q = np.array([[-3.06 ,  1.005 , 1.01 ],
 [ 1.005, -3.065,  1.015],
 [ 1.01,   1.015, -3.09 ]])

def qubo_to_ising(Q):
    n = len(Q)
    J = np.zeros((n, n))
    h = np.zeros(n)
    offset = 0
    for i in range(n):
        for j in range(n):
            if i == j:
                h[i] += Q[i][j] / 2 
                offset += Q[i][j] / 4
            else:
                J[i][j] += Q[i][j] / 4
                h[i] += Q[i][j] / 4
                h[j] += Q[i][j] / 4
                offset += Q[i][j] / 4
            
    return J, h, offset
    
J_matrix, h_vector, final_offset = qubo_to_ising(Q)

print("Linear Fields (h):")
print(h_vector)
print("\nCouplings (J):")
print(J_matrix)
print("\nConstant Offset:")
print(final_offset)
```
